```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import matplotlib as mpl
import seaborn as sb
mpl.rcParams['figure.dpi'] = 400
df = pd.read_csv('chapter1_cleand_data.csv')
```

```python
features_response = df.columns.tolist()
features_response
```

```python
items_to_remove = ['Unnamed: 0', 'ID','SEX','PAY_2', 'PAY_3', 'PAY_4', 'PAY_5', 'PAY_6','EDUCATION_CAT', 'graduate school', 'high school', 'others','university']
```

```python
features_response = [item for item in features_response if item not in items_to_remove]
features_response
```

```python
X= df[features_response].iloc[:,2]
X
Y= df[features_response].iloc[:,2]
Y
```

```python
marriage_one = Y == 1
marriage_two = Y == 2
marriage_three = Y == 3
```

```python
axes = plt.axes()
axes.hist(df.loc[marriage_one, 'AGE'], alpha = 0.5, color = 'blue')
axes.hist(df.loc[marriage_two, 'AGE'], alpha = 0.5, color = 'yellow')


axes.hist(df.loc[marriage_three, 'AGE'], alpha = 0.5,color = 'red')
axes.tick_params(axis = 'x', labelrotation = 45)

axes.set_xlabel('Age')
axes.set_ylabel('Frequency')
axes.legend(['Married', 'Not Married','unknown'])
axes.set_title('Age by marital status')
```

```python
bin_edges= list(range(20,80,5))
axes = plt.axes()
axes.hist(df.loc[marriage_one, 'AGE'], bins = bin_edges, density = True, alpha = 0.5, color = 'blue')
axes.hist(df.loc[marriage_two, 'AGE'], bins = bin_edges, density = True, alpha = 0.5, color = 'yellow')
axes.hist(df.loc[marriage_three, 'AGE'],bins = bin_edges, density = True, alpha = 0.5, color = 'red')
axes.tick_params(axis = 'x', labelrotation  = 45)
axes.set_xlabel('Age')
axes.set_ylabel('Proporation of Frequency')
y_ticks = axes.get_yticks()
axes.set_yticklabels(np.round(y_ticks*5, 2))
axes.legend(['Married', 'Not Married', 'Unknown'])
axes.set_title('Normalized distributions of Age by marital status')
```
